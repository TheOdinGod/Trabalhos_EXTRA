# **4. Análise Exploratória dos Dados**

```python
# 1.Verificando dados únicos
(df['Usuário'].unique())
```

```python
# 2.Verificando dados únicos
print(df['Micro'].unique())
```

```python
# 3.Verificando dados únicos e quantidade
print(df['Macro'].value_counts())
```

```python
# 4.Distribuição de Idade das Aves: 7, 14, 21, 28, 35, 42
plt.figure(figsize=(7,3)) # tamanho da figura
# Bins nas idades específicas: finalizando com 49 dias
sns.histplot(df['Idade'], bins=[7, 14, 21, 28, 35, 42, 49], kde=True, color='green')
plt.title('Distribuição da Idade das Aves') # titulo da figura
plt.xlabel('Idade (dias)') # idade em dias
plt.ylabel('Quantidade') # quantidade escala numérica
# Idades desejadas: de 7 em 7 dias
plt.xticks([7, 14, 21, 28, 35, 42])
plt.grid(True)
plt.show()
```

```python
# 5.Idades e Necropsias
idades = [7, 14, 21, 28, 35, 42] # de 7 em 7 dias
necropsias = [1, 2, 5, 6, 9, 12] # intervalos de quantidade
plt.figure(figsize=(9,5)) # tamanho da figura
plt.plot(idades, necropsias, marker='X', linestyle='--', color='green')
plt.title('Número de Aves Necropsiadas por Idade')
plt.xlabel('Idade das Aves (dias)')
plt.ylabel('Número de Aves Necropsiadas')
plt.xticks(idades)
plt.ylim(1, 15)  # de 1 a 15 aves
plt.grid(True)
plt.show()
```

```python
# Gráfico de Barras para Condições Anormais
idades = [7, 14, 21, 28, 35, 42]
espuma = [2, 4, 6, 8, 10, 12]       # Crescendo de 2 em 2
caseo = [1, 3, 5, 7, 9, 11]         # Crescendo de 2 em 2 começando em 1
septcemia = [1, 2, 3, 4, 5, 6]      # Crescendo de 1 em 1
# Posição das barras
largura = 0.25
x = np.arange(len(idades))
plt.figure(figsize=(6,3))
# As barras
plt.bar(x - largura, espuma, width=largura, label='Espuma', color='gray')
plt.bar(x, caseo, width=largura, label='Cáseo', color='yellow')
plt.bar(x + largura, septcemia, width=largura, label='Septcemia', color='orange')
# Personalizações do gráfico
plt.title('Condições Anormais por Idade das Aves (0 a 42 dias)')
plt.xlabel('Idade (dias)')
plt.ylabel('Número de Aves (Necropsiadas)')
plt.xticks(x, idades) # idades
plt.ylim(0, 15)  # 0 a 15 aves
plt.grid(axis='y')
# Legenda
plt.legend()
plt.show()
```

# **6. Análise visual dos dados**

```python
# Gráfico de Pizza para 'Macro'
plt.figure(figsize=(7, 5))
df['Macro'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribuição das Macros (Porcentagem de Necropsias)')
plt.ylabel('')  # deixar eixo y sem nome
plt.show()
```

```python
# Gráfico de Barras para 'Micro'
plt.figure(figsize=(8, 6))
df['Micro'].value_counts().plot(kind='bar', color='purple')
plt.title('Distribuição das Micros (Número de Necropsias)')
plt.xlabel('Anormalidades')
plt.ylabel('Número de Necropsias')
plt.xticks(rotation=45, ha='right')  # Rotacionar rótulos do eixo x para melhor visualização
plt.tight_layout() # Ajusta o layout para evitar sobreposição
plt.show()
```

```python
# Gráfico de barras para 'Macro' e suas 'Micros' correspondentes
macro_micro_counts = df.groupby(['Macro', 'Micro']).size().unstack(fill_value=0)
macro_micro_counts.plot(kind='bar', stacked=True, figsize=(9, 5), colormap='turbo')
plt.title('Distribuição de Micros por Macro')
plt.xlabel('MACROS')
plt.ylabel('Número de Necropsias')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title='Micro', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
